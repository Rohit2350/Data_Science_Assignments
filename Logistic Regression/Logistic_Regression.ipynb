# Logistic Regression

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from pandas.plotting import scatter_matrix
from sklearn.metrics import classification_report,accuracy_score
```

***

```python
df1=pd.read_csv("bank-full.csv",delimiter=";")
df1
```

```python
df1.info()
```

```python
data1=pd.get_dummies(df1,columns=["job","marital","education","contact","poutcome","month"])
data1
```

```python
pd.set_option("Display.max.columns",None)
data1
```

```python
data1.info()
```

```python
data1['default'] = np.where(data1['default'].str.contains("yes"), 1, 0)
data1['housing'] = np.where(data1['housing'].str.contains("yes"), 1, 0)
data1['loan'] = np.where(data1['loan'].str.contains("yes"), 1, 0)
data1['y'] = np.where(data1['y'].str.contains("yes"), 1, 0)
data1
```

```python
data1.info()
```

```python
x=pd.concat([data1.iloc[:,0:10],data1.iloc[:,11:]],axis=1)
y=data1.iloc[:,10]
```

```python
data1.hist(bins=40,figsize=(20,15))
plt.show
```

```python
columns = ['age','default','housing','loan','day','duration']
plt.figure(figsize = (15, 20))
plotnumber = 1
for i in range(len(columns)):
    if plotnumber <= 10:
        ax = plt.subplot(5, 2, plotnumber)
        sns.countplot(x = columns[i],ax=ax, data =data1)
    plotnumber += 1
    plt.tight_layout()
plt.show()
```

```python
classifier=LogisticRegression()
classifier.fit(x,y)
```

```python
y_pred=classifier.predict(x)
y_pred
```

```python
y_pred_df=pd.DataFrame({'actual_y':y,'y_pred_prob':y_pred})
y_pred_df
```

```python
confusion_matrix = confusion_matrix(y,y_pred)
confusion_matrix
```

```python
classifier.predict_proba(x)[:,1] 
```

```python
fpr,tpr,thresholds=roc_curve(y,classifier.predict_proba(x)[:,1])
plt.plot(fpr,tpr,color='red')
auc=roc_auc_score(y,y_pred)

plt.plot(fpr,tpr,color='red',label='logit model(area  = %0.2f)'%auc)
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')
plt.show()

print('auc accuracy:',auc)
```
